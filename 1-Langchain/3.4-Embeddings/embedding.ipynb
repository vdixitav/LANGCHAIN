```python
import os
from dotenv import load_dotenv
load_dotenv() # load all envt variables
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text='This is tutorial on open ai embedding'
query_result=embeddings.embed_query(text)
query_result
```

```python
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
embeddings_1024
```

```python
text='This is tutorial on open ai embedding'
query_result=embeddings_1024.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
# loading txt doc
from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt")
text_documents=loader.load() 
text_documents
```

```python
# converting into text

from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500 ,chunk_overlap=50)
final_documents=text_splitter.split_documents(text_documents)
final_documents
```

```python
# vector embedding and vector storeDB

from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_documents,embeddings_1024)
db

```

```python
query='Speech is the use of the human voice as a medium for language'
retrived_result=db.similarity_search(query)
print(retrived_result)
```

```python
query='myname usdfjbsdbfjksbdjcvusvdfusgdkjcbdsis dixitha'
retrived_result=db.similarity_search(query)
print(retrived_result)
```
